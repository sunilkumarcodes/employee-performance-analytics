```python
fig,ax=plt.subplots(2,2,figsize=(10,10))
plt.suptitle("Understanding the distribution of various factors",fontsize=20)

sns.histplot(hr['Age'], ax=ax[0,0])
ax[0][0].set_title('Distribution of age')

sns.histplot(hr['TotalWorkingYears'], ax=ax[0,1])
ax[0][1].set_title('Distribution of Total Working Years ')

sns.histplot(hr['YearsAtCompany'], ax=ax[1,0])
ax[1][0].set_title('Distribution of the years at company')

sns.histplot(hr['YearsInCurrentRole'], ax=ax[1,1])
ax[1][1].set_title('distribution of years in current role')
plt.show()


```

From the plots ,we find that majority of the employees are in the age group of 35-40 years and have a total experience of around 8-10 years.They have typically spent around 2-4 years at years at the current organization and spent 2-3 years in the current role they are in.

```python
sns.countplot(hr['Attrition'])
plt.show()
```

Above plot shows that around 250 people have left the organization while the organization has been able to retain around 1200 people.

```python
from numpy import median
sns.barplot(x='Attrition',y='MonthlyIncome',hue='Gender',data=hr,estimator=median)
plt.show()
```

```python
from numpy import median
sns.barplot(x='Attrition', y='MonthlyIncome',hue='JobSatisfaction',data=hr,estimator=median)
plt.show()
```

```python
f,ax=plt.subplots(figsize=(15,10))
sns.boxplot(x='JobSatisfaction',y='MonthlyIncome',data=hr, hue='JobLevel',palette='Set3')
plt.legend(loc='best')
plt.show()
```

We can clearly observe the difference in monthly income across different job levels .The difference in the monthly income is observed by the median value for different job satisfaction level.

```python
sns.violinplot(x="Attrition",y="YearsAtCompany",hue="Gender",data=hr, palette="muted", split=True, inner="stick")
plt.show()
```

Here, We have plotted the number of years spent in an organization based on gender. The middle dashed line shows the median. the lines above and  below the median show the interquartile range. The denser part shows the maximum population falls under that range and thinner part shows the lesser population.

```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.jointplot(x='Age', y='MonthlyIncome', data=hr, kind='scatter')


plt.show()
```

Scatter plot shows the relatuonship between Age and Monthly income.We Can find  linear relationship. Further, the density plot above shows the distribution of age while density plot in the right shows the distribution of the monthly income.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


hr['age_group'] = pd.cut(hr['Age'], 3, labels=['Young', 'Middle', 'Senior'])


sns.catplot(x='Attrition', y='MonthlyIncome', hue='JobLevel', col='age_group', col_wrap=2, kind='box', data=hr)


plt.show()
```

Above factor plot shows that monthly income plays an important role in retaining the employees in an organization. it can be observed across job levels and different age groups.

```python
import seaborn as sns
import matplotlib.pyplot as plt


g = sns.FacetGrid(hr, col="JobSatisfaction", row="Gender")
g.map(sns.kdeplot, "MonthlyIncome", "YearsInCurrentRole", fill=True)


plt.show()
```

```python
data1=['Attrition', 'Age', 'MonthlyIncome','DistanceFromHome']
sns.pairplot(hr[data1], kind="reg", diag_kind="kde",hue='Attrition')
plt.show()
```

Pairwise plots beteen continuous variables show the relationship between them. for example observing the relationship between Age and Monthly income , we can find that age,monthly income has increased but the increase is not similar for both groups(Attrition and Retention).

```python
data2=['Gender','HourlyRate','MonthlyRate','PercentSalaryHike']
sns.pairplot(hr[data2], kind="reg", diag_kind="kde", hue='Gender')
plt.show()
```

The above plot does not convey much of any relationship between variables across gender. This shows that hourly rate, daily rate,monthly rate and percentage salary hike is same foe both female and male employees.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Ensure your DataFrame is loaded
# hr = pd.read_csv('path_to_your_file.csv')

# Selecting only numeric columns for correlation
numeric_cols = hr.select_dtypes(include=['float64', 'int64'])

# Create the heatmap
f, ax = plt.subplots(figsize=(18, 18))
sns.heatmap(numeric_cols.corr(), annot=True, linewidths=.4, fmt='.1f', ax=ax)

# Display the plot
plt.show()
```

Above plot are the correlation map for all numeric variables.

Two variables are said to be highly correlation when they have a value of 0.7 or grater. The correlation plot between all continuous variables indicate that years company and year with current manager,years in current role and years with current manager,monthly income and total working years,age and total working years,percent salary hike and performance rating are highly correlated.
